```python
# Core imports
from pathlib import Path
import warnings

import numpy as np
import pandas as pd

# Optional visualization stack for the dashboard
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots


warnings.filterwarnings("ignore")

# Path cấu hình
BASE_DIR = Path(r"d:\Vscode\Self\Data Science\Book Notes\Data Visualization\final\Final_Data_Viz")
DATA_DIR = BASE_DIR / "data" / "main" / "raw"

# Helper load
def load_csv(name, parse_dates=None):
    path = DATA_DIR / name
    return pd.read_csv(path, parse_dates=parse_dates)

# Load all datasets
dfs = {
    "products": load_csv("products.csv"),
    "customers": load_csv("customers.csv", parse_dates=["signup_date"]),
    "promotions": load_csv("promotions.csv", parse_dates=["start_date", "end_date"]),
    "geography": load_csv("geography.csv"),
    "orders": load_csv("orders.csv", parse_dates=["order_date"]),
    "order_items": load_csv("order_items.csv"),
    "payments": load_csv("payments.csv"),
    "shipments": load_csv("shipments.csv", parse_dates=["ship_date", "delivery_date"]),
    "returns": load_csv("returns.csv", parse_dates=["return_date"]),
    "reviews": load_csv("reviews.csv", parse_dates=["review_date"]),
    "sales": load_csv("sales.csv", parse_dates=["Date"]),
    "inventory": load_csv("inventory.csv", parse_dates=["snapshot_date"]),
    "web_traffic": load_csv("web_traffic.csv", parse_dates=["date"]),
}

# Basic cleanup
def standardize_columns(df):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    return df

for key in dfs:
    dfs[key] = standardize_columns(dfs[key])

# Ensure expected numeric columns are numeric
numeric_columns = {
    "products": ["price", "cogs"],
    "customers": ["zip"],
    "promotions": ["discount_value", "stackable_flag", "min_order_value"],
    "geography": ["zip"],
    "orders": ["order_id", "customer_id", "zip"],
    "order_items": ["order_id", "product_id", "quantity", "unit_price", "discount_amount"],
    "payments": ["order_id", "payment_value", "installments"],
    "shipments": ["order_id", "shipping_fee"],
    "returns": ["order_id", "product_id", "return_quantity", "refund_amount"],
    "reviews": ["order_id", "product_id", "customer_id", "rating"],
    "sales": ["Revenue", "COGS"],
    "inventory": ["product_id", "stock_on_hand", "units_received", "units_sold", "stockout_days", "days_of_supply", "fill_rate", "stockout_flag", "overstock_flag", "reorder_flag", "sell_through_rate", "year", "month"],
    "web_traffic": ["sessions", "unique_visitors", "page_views", "bounce_rate", "avg_session_duration_sec"],
}

for name, cols in numeric_columns.items():
    for col in cols:
        if col in dfs[name].columns:
            dfs[name][col] = pd.to_numeric(dfs[name][col], errors="coerce")

# Quick sanity checks
for name, df in dfs.items():
    print(f"{name:12s} -> rows: {len(df):>8,} | cols: {df.shape[1]:>3}")
```

## 💎 PHẦN 1: CÁC CHỈ SỐ TÀI CHÍNH CỐT LÕI (BIG NUMBER KPIs)

### 🎯 1. Mục tiêu của phần này
Phần này được thiết kế để cung cấp một cái nhìn tức thời về **Sức khỏe sinh tồn** của doanh nghiệp.  
Thay vì chỉ quan sát doanh thu đơn thuần, lớp KPI này giúp ban điều hành nắm nhanh: doanh nghiệp đang tăng trưởng thật sự hay chỉ đang tạo ra quy mô bề ngoài nhưng tiềm ẩn rủi ro về lợi nhuận.

### 🎨 2. Logic màu sắc Blue-to-Red
Hệ màu được xây dựng theo nguyên tắc **Blue-to-Red Diverging Scale** nhằm hỗ trợ ra quyết định nhanh:

- **Xanh dương (Blue):** Tín hiệu tích cực, an toàn, vượt ngưỡng hoặc đang vận hành tốt.
- **Đỏ (Red):** Tín hiệu cảnh báo, cho thấy chỉ số đang dưới kỳ vọng hoặc cần can thiệp sớm.

Cách phối màu này giúp người xem nhận ra ngay tình trạng sức khỏe tài chính mà không cần đọc quá nhiều số liệu chi tiết.

---

### 📊 3. Bốn chỉ số chiến lược cần theo dõi

| Chỉ số | Ý nghĩa quản trị | Vai trò trong ra quyết định |
| :--- | :--- | :--- |
| **Tổng Doanh thu** | Tổng giá trị hóa đơn phát sinh trong kỳ. Đây là chỉ số **Top-line** phản ánh quy mô thị trường và tốc độ mở rộng của doanh nghiệp - khả năng chiếm lĩnh thị trường của doanh nghiệp . | Cho biết doanh nghiệp đang bán được bao nhiêu và đang mở rộng nhanh hay chậm. |
| **Lợi nhuận Thuần (Net Profit)** | Phần giá trị thực tế còn lại sau khi trừ **Giá vốn** và **Thất thoát do trả hàng**. Đây là chỉ số quan trọng nhất vì phản ánh khả năng tạo ra giá trị thật. | Quyết định trực tiếp đến sức khỏe tài chính và khả năng duy trì hoạt động dài hạn. |
| **Biên Lợi nhuận Gộp (%)** | Phản ánh hiệu quả của nguồn hàng, cấu trúc giá bán và chiến lược định giá. Phản ánh khả năng cạnh tranh của doanh nghiệp. **Công thức:** Lợi nhuận gộp / Doanh thu. | Đánh giá mức độ “dày” của lợi nhuận trước các chi phí vận hành và marketing. |
| **Tỷ lệ Trả hàng (%)** | Chỉ số đặc thù của E-commerce thời trang, phản ánh mức độ hài lòng về sản phẩm, độ phù hợp của size, và rủi ro vận hành. | Nếu quá cao, lợi nhuận thực tế sẽ bị bào mòn dù doanh thu nhìn bên ngoài vẫn tăng. |

### 🧠 4. Vì sao chọn đúng 4 chỉ số này?
Bộ 4 KPI này được thiết kế để cân bằng giữa:

- **Chỉ số Tấn công:** Tổng Doanh thu, đại diện cho khả năng mở rộng quy mô, chiếm lĩnh thị trường.
- **Chỉ số Phòng thủ:** Lợi nhuận Thuần, Biên Lợi nhuận Gộp và Tỷ lệ Trả hàng, đại diện cho sự ổn định, chất lượng tăng trưởng, khả năng cạnh tranh.

Sự kết hợp này giúp tránh cái bẫy **“tăng trưởng ảo”**: doanh thu tăng nhưng lợi nhuận giảm, hoặc bán được nhiều nhưng tỷ lệ trả hàng cao khiến hiệu quả thực tế suy yếu.

### ✅ Kết luận ngắn
Bộ Big Number KPIs không chỉ nhằm hiển thị số liệu lớn, mà còn là lớp cảnh báo sớm cho ban điều hành.  
Nếu Doanh thu tăng nhưng Lợi nhuận Thuần không cải thiện, hoặc Tỷ lệ Trả hàng vượt ngưỡng, đó là dấu hiệu cho thấy cần xem lại chiến lược sản phẩm, giá bán, khuyến mãi và vận hành.

```python
# Nếu cell trước đang load vào dict dfs, tạo alias đúng theo yêu cầu
df_sales = dfs["sales"].copy()
df_order_items = dfs["order_items"].copy()
df_returns = dfs["returns"].copy()

# 1) Thiết lập ngưỡng
nguong_bien_loi_nhuan = 0.30
nguong_ty_le_tra_hang = 0.15

# 2) Tính toán các chỉ số
tong_doanh_thu = df_sales["Revenue"].sum()
tong_gia_von = df_sales["COGS"].sum()

# Tính giá trị trả hàng
if "order_item_id" in df_returns.columns and "order_item_id" in df_order_items.columns:
    df_tra_hang = df_returns.merge(
        df_order_items,
        on="order_item_id",
        how="left",
        suffixes=("_return", "_item")
    )

    if {"quantity", "unit_price"}.issubset(df_tra_hang.columns):
        gia_tri_tra_hang = (df_tra_hang["quantity"] * df_tra_hang["unit_price"]).sum()
    else:
        gia_tri_tra_hang = df_tra_hang["unit_price"].sum()
else:
    # Fallback an toàn nếu không có order_item_id
    if {"quantity", "unit_price"}.issubset(df_order_items.columns):
        gia_tri_tra_hang = df_returns.shape[0] * df_order_items["unit_price"].mean()
    else:
        gia_tri_tra_hang = df_returns.shape[0]

loi_nhuan_thuan = tong_doanh_thu - tong_gia_von - gia_tri_tra_hang
bien_loi_nhuan_gop = (tong_doanh_thu - tong_gia_von) / tong_doanh_thu if tong_doanh_thu != 0 else np.nan
ty_le_tra_hang = len(df_returns) / len(df_order_items) if len(df_order_items) != 0 else np.nan

# 3) Logic màu sắc
def lay_mau_chi_so(gia_tri, nguong, dao_nguoc=False):
    if pd.isna(gia_tri):
        return "#808080"
    if dao_nguoc:
        return "#0000FF" if gia_tri <= nguong else "#FF0000"
    return "#0000FF" if gia_tri >= nguong else "#FF0000"

mau_doanh_thu = lay_mau_chi_so(tong_doanh_thu, 0)
mau_loi_nhuan = lay_mau_chi_so(loi_nhuan_thuan, 0)
mau_bien_loi_nhuan = lay_mau_chi_so(bien_loi_nhuan_gop, nguong_bien_loi_nhuan)
mau_ty_le_tra_hang = lay_mau_chi_so(ty_le_tra_hang, nguong_ty_le_tra_hang, dao_nguoc=True)

# 4) Hàm tạo indicator
def tao_indicator(tieu_de, gia_tri, mau_so, prefix="", suffix="", valueformat=","):
    return go.Indicator(
        mode="number",
        value=gia_tri,
        title={"text": f"<b>{tieu_de}</b>"},
        number={
            "prefix": prefix,
            "suffix": suffix,
            "valueformat": valueformat,
            "font": {"size": 28, "color": mau_so},
        },
    )

# Chuẩn hóa hiển thị
doan_thu_hien_thi = tong_doanh_thu
loi_nhuan_hien_thi = loi_nhuan_thuan
bien_loi_nhuan_hien_thi = bien_loi_nhuan_gop * 100 if pd.notna(bien_loi_nhuan_gop) else np.nan
ty_le_tra_hang_hien_thi = ty_le_tra_hang * 100 if pd.notna(ty_le_tra_hang) else np.nan

# 5) Tạo 4 thẻ KPI dạng Indicator
fig = make_subplots(
    rows=1,
    cols=4,
    specs=[[{"type": "indicator"}, {"type": "indicator"}, {"type": "indicator"}, {"type": "indicator"}]],
    horizontal_spacing=0.08
)

fig.add_trace(
    tao_indicator("TỔNG DOANH THU", doan_thu_hien_thi, mau_doanh_thu, suffix="VNĐ ", valueformat=",.0f"),
    row=1, col=1
)

fig.add_trace(
    tao_indicator("LỢI NHUẬN THUẦN", loi_nhuan_hien_thi, mau_loi_nhuan, suffix="VNĐ ", valueformat=",.0f"),
    row=1, col=2
)

fig.add_trace(
    tao_indicator("BIÊN LỢI NHUẬN GỘP", bien_loi_nhuan_hien_thi, mau_bien_loi_nhuan, suffix="%", valueformat=".2f"),
    row=1, col=3
)

fig.add_trace(
    tao_indicator("TỶ LỆ TRẢ HÀNG", ty_le_tra_hang_hien_thi, mau_ty_le_tra_hang, suffix="%", valueformat=".2f"),
    row=1, col=4
)

fig.update_layout(
    template="plotly_white",
    height=260,
    margin=dict(l=20, r=20, t=30, b=10),
)

fig.show()
```

```python
# 6) In thống kê các chỉ số để phục vụ phân tích insight
def dinh_dang_tien_te(x):
    return f"{x:,.0f} VNĐ"

def dinh_dang_ty_le(x):
    return f"{x:.2%}"

trang_thai_bien_loi_nhuan = "TỐT" if bien_loi_nhuan_gop >= nguong_bien_loi_nhuan else "CẦN CẢI THIỆN"
trang_thai_ty_le_tra_hang = "TỐT" if ty_le_tra_hang <= nguong_ty_le_tra_hang else "CẦN CẢI THIỆN"

bang_thong_ke_kpi = pd.DataFrame([
    {
        "Chỉ số": "Tổng doanh thu",
        "Giá trị": dinh_dang_tien_te(tong_doanh_thu),
        "Ngưỡng": "-",
        "Trạng thái": "CƠ SỞ"
    },
    {
        "Chỉ số": "Tổng giá vốn",
        "Giá trị": dinh_dang_tien_te(tong_gia_von),
        "Ngưỡng": "-",
        "Trạng thái": "CƠ SỞ"
    },
    {
        "Chỉ số": "Giá trị trả hàng",
        "Giá trị": dinh_dang_tien_te(gia_tri_tra_hang),
        "Ngưỡng": "-",
        "Trạng thái": "CƠ SỞ"
    },
    {
        "Chỉ số": "Lợi nhuận thuần",
        "Giá trị": dinh_dang_tien_te(loi_nhuan_thuan),
        "Ngưỡng": "-",
        "Trạng thái": "CƠ SỞ"
    },
    {
        "Chỉ số": "Biên lợi nhuận gộp",
        "Giá trị": dinh_dang_ty_le(bien_loi_nhuan_gop),
        "Ngưỡng": dinh_dang_ty_le(nguong_bien_loi_nhuan),
        "Trạng thái": trang_thai_bien_loi_nhuan
    },
    {
        "Chỉ số": "Tỷ lệ trả hàng",
        "Giá trị": dinh_dang_ty_le(ty_le_tra_hang),
        "Ngưỡng": dinh_dang_ty_le(nguong_ty_le_tra_hang),
        "Trạng thái": trang_thai_ty_le_tra_hang
    }
])

print("\n===== THỐNG KÊ CÁC CHỈ SỐ KPI =====")
print(bang_thong_ke_kpi.to_string(index=False))

print("\n===== TÓM TẮT NHANH =====")
print(f"- Tổng doanh thu: {dinh_dang_tien_te(tong_doanh_thu)}")
print(f"- Tổng giá vốn: {dinh_dang_tien_te(tong_gia_von)}")
print(f"- Giá trị trả hàng: {dinh_dang_tien_te(gia_tri_tra_hang)}")
print(f"- Lợi nhuận thuần: {dinh_dang_tien_te(loi_nhuan_thuan)}")
print(f"- Biên lợi nhuận gộp: {dinh_dang_ty_le(bien_loi_nhuan_gop)} | Ngưỡng: {dinh_dang_ty_le(nguong_bien_loi_nhuan)} | {trang_thai_bien_loi_nhuan}")
print(f"- Tỷ lệ trả hàng: {dinh_dang_ty_le(ty_le_tra_hang)} | Ngưỡng: {dinh_dang_ty_le(nguong_ty_le_tra_hang)} | {trang_thai_ty_le_tra_hang}")
```

### Nhận xét

- **Hiệu quả tài chính tổng thể:**

## 📈 PHẦN 2: PHÂN TÍCH XU HƯỚNG BIẾN ĐỘNG DOANH THU & GIÁ VỐN

### 🎯 1. Mục tiêu của phần này
Nếu Phần 1 cho ta cái nhìn **tĩnh** về sức khỏe sinh tồn của doanh nghiệp qua các KPI tổng số, thì Phần 2 chuyển sang góc nhìn **động** theo chuỗi thời gian.

Mục tiêu trọng tâm là quan sát **Khoảng hở Lợi nhuận (Profit Margin Gap)** giữa **Doanh thu** và **Giá vốn (COGS)** theo thời gian, từ đó nhận diện:
- Thời điểm kinh doanh hiệu quả, khi doanh thu tăng nhanh hơn giá vốn.
- Thời điểm rò rỉ lợi nhuận, khi giá vốn tiến sát hoặc bào mòn doanh thu.
- Các giai đoạn có dấu hiệu bất thường cần phân tích sâu hơn về vận hành, giá bán hoặc khuyến mãi.

---

### 🛠️ 2. Kỹ thuật xử lý dữ liệu & hiển thị

#### 📉 Làm mượt dữ liệu bằng Moving Average (MA7)
Dữ liệu doanh thu và giá vốn theo ngày thường có nhiều biến động ngắn hạn do:
- Khác biệt giữa ngày thường và cuối tuần.
- Biến động hành vi mua sắm theo từng thời điểm.
- Các “răng cưa” ngẫu nhiên trong giao dịch hằng ngày.

Vì vậy, sử dụng **đường trung bình động 7 ngày (MA7)** là một lựa chọn phù hợp để:
- Triệt tiêu nhiễu ngắn hạn.
- Làm rõ **xu hướng thực (True Trend)**.
- Giúp nhà quản trị nhìn ra chuyển động nền thay vì bị cuốn theo dao động nhỏ.

#### 🧩 Biểu đồ Miền & Đường (Area & Line Layering)
Phần này nên kết hợp hai lớp trực quan:

- **Area Chart cho Doanh thu**: thể hiện quy mô dòng tiền đổ vào, tạo cảm giác “đầy” và trực quan về sức mua.
- **Line Chart cho Giá vốn**: thể hiện dòng chi phí đầu vào, giúp so sánh trực tiếp với doanh thu.

Cách layering này tạo ra tương phản thị giác rõ rệt, từ đó làm nổi bật **biên lợi nhuận** và giúp người xem nhanh chóng thấy được giai đoạn nào doanh nghiệp đang tạo giá trị tốt.

#### 🎛️ Tính tương tác bằng Range Slider
Biểu đồ nên tích hợp **Range Slider** để người dùng có thể:
- Chọn nhanh một giai đoạn cụ thể.
- Zoom vào các năm có biến động lớn.
- Phân tích sâu các thời điểm đặc biệt như mùa Tết, chiến dịch khuyến mãi hoặc giai đoạn suy giảm.

Tính năng này đặc biệt hữu ích khi dashboard cần phục vụ đồng thời cả quan sát tổng quan lẫn đào sâu phân tích.

---

### 🎨 3. Màu sắc (Color Strategy)

#### 🔵 Xanh dương cho Doanh thu
**Xanh dương** đại diện cho:
- Dòng tiền đổ vào.
- Sự tăng trưởng.
- Năng lực mở rộng quy mô kinh doanh.

Đây là màu chủ đạo của phần doanh thu để nhấn mạnh vai trò “đầu vào tích cực” của hệ thống kinh doanh.

#### 🔴 Đỏ cho Giá vốn
**Đỏ** đại diện cho:
- Dòng chi phí đổ ra.
- Áp lực lên biên lợi nhuận.
- Rủi ro vận hành hoặc tối ưu chi phí chưa hiệu quả.

Khi đường **Đỏ** tiến gần sát vùng miền **Xanh**, đó là tín hiệu cảnh báo rằng:
- Chi phí hàng bán đang tăng mạnh.
- Biên lợi nhuận bị thu hẹp.
- Doanh nghiệp cần xem lại chiến lược giá, sourcing, hoặc cấu trúc khuyến mãi.

---

### 🌸 4. Phân tích tính mùa vụ (Seasonality & Event Context)

Một trong những giá trị lớn nhất của biểu đồ này là khả năng đánh dấu các mốc **Lễ hội mua sắm Tết hằng năm** và các sự kiện kinh doanh quan trọng.

Việc gắn nhãn mùa vụ giúp trả lời câu hỏi ở cấp độ chẩn đoán:

- **Tại sao doanh thu tăng đột biến?**
- Đây là tăng trưởng thật hay chỉ là hiệu ứng mùa vụ?
- Giá vốn có tăng cùng nhịp hay đang làm giảm hiệu quả lợi nhuận?

Đây là bước chuyển từ **Level 1: Descriptive** sang **Level 2: Diagnostic**, đồng thời chuẩn bị nền cho **Level 3: Predictive** khi doanh nghiệp cần dự báo trước mùa cao điểm tiếp theo.

---

### 🧭 5. Cách đọc và nhận xét biểu đồ

Khi quan sát biểu đồ, người xem nên tập trung vào hai tín hiệu chính:

#### 📐 Độ dốc của đường cong
- Độ dốc đi lên cho thấy tăng trưởng đang diễn ra.
- Độ dốc phẳng hoặc đi xuống cho thấy xu hướng chậm lại hoặc suy giảm.
- So sánh độ dốc của Doanh thu và Giá vốn giúp đánh giá chất lượng tăng trưởng.

#### 🕳️ Độ rộng của khoảng hở
- Khoảng hở càng rộng, biên lợi nhuận càng khỏe.
- Khoảng hở thu hẹp dần là dấu hiệu cho thấy lợi nhuận đang bị bào mòn.
- Nếu Giá vốn tăng nhanh hơn Doanh thu, doanh nghiệp có nguy cơ rơi vào tình trạng tăng trưởng doanh thu nhưng suy giảm hiệu quả.

```python
# 1) Tiền xử lý dữ liệu
df_trend = (
    df_sales.copy()
    .groupby("Date", as_index=False)[["Revenue", "COGS"]]
    .sum()
    .sort_values("Date")
)

# Làm mượt MA7
df_trend["Doanh thu (MA7)"] = df_trend["Revenue"].rolling(window=7, min_periods=1).mean()
df_trend["Giá vốn (MA7)"] = df_trend["COGS"].rolling(window=7, min_periods=1).mean()
df_trend["Gross Profit (MA7)"] = df_trend["Doanh thu (MA7)"] - df_trend["Giá vốn (MA7)"]

# 2) Xác định khoảng thời gian Tết Nguyên Đán gần cuối tháng 1 - đầu tháng 2
tet_ranges = []
for year in sorted(df_trend["Date"].dt.year.unique()):
    start = pd.Timestamp(f"{year}-01-20")
    end = pd.Timestamp(f"{year}-02-10")

    mask = (df_trend["Date"] >= start) & (df_trend["Date"] <= end)
    if mask.any():
        tet_ranges.append((df_trend.loc[mask, "Date"].min(), df_trend.loc[mask, "Date"].max()))

# 3) Hàm dùng chung để vẽ vùng Tết
def them_mua_tet(fig):
    for start_date, end_date in tet_ranges:
        fig.add_vrect(
            x0=start_date,
            x1=end_date,
            fillcolor="rgba(255, 223, 0, 0.18)",
            line_width=0,
            layer="below",
            annotation_text="Mùa sắm Tết",
            annotation_position="top left",
        )

# 4) BIỂU ĐỒ DOANH THU THEO THỜI GIAN
fig_rev = go.Figure()

fig_rev.add_trace(
    go.Scatter(
        x=df_trend["Date"],
        y=df_trend["Doanh thu (MA7)"],
        mode="lines",
        name="Doanh thu (MA7)",
        line=dict(color="rgba(30, 144, 255, 1)", width=2.8),
        fill="tozeroy",
        fillcolor="rgba(173, 216, 230, 0.35)",
        hovertemplate="%{x|%d/%m/%Y}<br>Doanh thu: %{y:,.0f} VNĐ<extra></extra>",
    )
)

them_mua_tet(fig_rev)

fig_rev.update_layout(
    title={
        "text": "DOANH THU THEO THỜI GIAN (LÀM MƯỢT 7 NGÀY)",
        "x": 0.5,
        "xanchor": "center",
    },
    template="plotly_white",
    width=1400,
    height=520,
    legend=dict(
        title="Chú thích",
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
    ),
    margin=dict(l=60, r=40, t=80, b=60),
)

fig_rev.update_xaxes(
    title_text="Thời gian",
    rangeslider=dict(visible=True),
    showgrid=False,
)

fig_rev.update_yaxes(
    title_text="Doanh thu (VNĐ)",
    tickformat=",.0f",
    showgrid=True,
    gridcolor="rgba(0,0,0,0.08)",
)

fig_rev.show()

# 5) BIỂU ĐỒ GIÁ VỐN THEO THỜI GIAN
fig_cogs = go.Figure()

fig_cogs.add_trace(
    go.Scatter(
        x=df_trend["Date"],
        y=df_trend["Giá vốn (MA7)"],
        mode="lines",
        name="Giá vốn (MA7)",
        line=dict(color="rgba(178, 34, 34, 1)", width=2.8),
        fill="tozeroy",
        fillcolor="rgba(255, 182, 193, 0.25)",
        hovertemplate="%{x|%d/%m/%Y}<br>Giá vốn: %{y:,.0f} VNĐ<extra></extra>",
    )
)

them_mua_tet(fig_cogs)

fig_cogs.update_layout(
    title={
        "text": "GIÁ VỐN THEO THỜI GIAN (LÀM MƯỢT 7 NGÀY)",
        "x": 0.5,
        "xanchor": "center",
    },
    template="plotly_white",
    width=1400,
    height=520,
    legend=dict(
        title="Chú thích",
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
    ),
    margin=dict(l=60, r=40, t=80, b=60),
)

fig_cogs.update_xaxes(
    title_text="Thời gian",
    rangeslider=dict(visible=True),
    showgrid=False,
)

fig_cogs.update_yaxes(
    title_text="Giá vốn (VNĐ)",
    tickformat=",.0f",
    showgrid=True,
    gridcolor="rgba(0,0,0,0.08)",
)

fig_cogs.show()

# 6) BIỂU ĐỒ LỢI NHUẬN GỘP THEO THỜI GIAN
# 6) BIỂU ĐỒ LỢI NHUẬN GỘP THEO THỜI GIAN
fig_gp = go.Figure()

df_trend["Gross Profit (MA7)"] = df_trend["Doanh thu (MA7)"] - df_trend["Giá vốn (MA7)"]

gross_profit_positive = df_trend["Gross Profit (MA7)"].where(df_trend["Gross Profit (MA7)"] >= 0, 0)
gross_profit_negative = df_trend["Gross Profit (MA7)"].where(df_trend["Gross Profit (MA7)"] < 0, 0)

# Phần dương: màu xanh dương
fig_gp.add_trace(
    go.Scatter(
        x=df_trend["Date"],
        y=gross_profit_positive,
        mode="lines",
        name="Lợi nhuận gộp dương",
        line=dict(color="rgba(30, 144, 255, 1)", width=2.8),
        fill="tozeroy",
        fillcolor="rgba(173, 216, 230, 0.30)",
        hovertemplate="%{x|%d/%m/%Y}<br>Lợi nhuận gộp: %{y:,.0f} VNĐ<extra></extra>",
    )
)

# Phần âm: màu đỏ
fig_gp.add_trace(
    go.Scatter(
        x=df_trend["Date"],
        y=gross_profit_negative,
        mode="lines",
        name="Lợi nhuận gộp âm",
        line=dict(color="rgba(178, 34, 34, 1)", width=2.8),
        fill="tozeroy",
        fillcolor="rgba(255, 182, 193, 0.35)",
        hovertemplate="%{x|%d/%m/%Y}<br>Lợi nhuận gộp: %{y:,.0f} VNĐ<extra></extra>",
    )
)

them_mua_tet(fig_gp)

fig_gp.add_hline(y=0, line_width=1, line_dash="dash", line_color="gray")

fig_gp.update_layout(
    title={
        "text": "LỢI NHUẬN GỘP THEO THỜI GIAN (DOANH THU - GIÁ VỐN, LÀM MƯỢT 7 NGÀY)",
        "x": 0.5,
        "xanchor": "center",
    },
    template="plotly_white",
    width=1400,
    height=520,
    legend=dict(
        title="Chú thích",
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
    ),
    margin=dict(l=60, r=40, t=80, b=60),
)

fig_gp.update_xaxes(
    title_text="Thời gian",
    rangeslider=dict(visible=True),
    showgrid=False,
)

fig_gp.update_yaxes(
    title_text="Lợi nhuận gộp (VNĐ)",
    tickformat=",.0f",
    showgrid=True,
    gridcolor="rgba(0,0,0,0.08)",
)

fig_gp.show()
```

```python
# =========================
# 1) Chuẩn bị dữ liệu ngày
# =========================
df_daily = (
    df_sales.copy()
    .groupby("Date", as_index=False)[["Revenue", "COGS"]]
    .sum()
    .sort_values("Date")
)

df_daily["Gross Profit"] = df_daily["Revenue"] - df_daily["COGS"]
df_daily["Year"] = df_daily["Date"].dt.year
df_daily["Month"] = df_daily["Date"].dt.month
df_daily["Quarter"] = df_daily["Date"].dt.quarter

month_map = {
    1: "Tháng 1", 2: "Tháng 2", 3: "Tháng 3", 4: "Tháng 4",
    5: "Tháng 5", 6: "Tháng 6", 7: "Tháng 7", 8: "Tháng 8",
    9: "Tháng 9", 10: "Tháng 10", 11: "Tháng 11", 12: "Tháng 12"
}

# =========================
# 2) Thống kê theo tháng
# =========================
monthly_summary = (
    df_daily.groupby(["Year", "Month"], as_index=False)
    .agg(
        Revenue=("Revenue", "sum"),
        COGS=("COGS", "sum"),
        Gross_Profit=("Gross Profit", "sum"),
    )
)

monthly_summary["Month Name"] = monthly_summary["Month"].map(month_map)

# Top 5 tháng có lợi nhuận gộp cao nhất theo từng năm
top5_months_each_year = (
    monthly_summary.sort_values(["Year", "Gross_Profit"], ascending=[True, False])
    .groupby("Year")
    .head(5)
    .reset_index(drop=True)
)

month_frequency_top5 = (
    top5_months_each_year.groupby(["Month", "Month Name"], as_index=False)
    .agg(
        Frequency_in_Top5=("Gross_Profit", "size"),
        Total_Gross_Profit=("Gross_Profit", "sum"),
    )
    .sort_values(["Frequency_in_Top5", "Total_Gross_Profit", "Month"], ascending=[False, False, True])
)

top3_months_most_frequent = month_frequency_top5.head(3)

# Các tháng có lợi nhuận gộp âm mạnh nhất
bottom_months_gross_profit = (
    monthly_summary.sort_values("Gross_Profit", ascending=True)
    .head(10)
    .reset_index(drop=True)
)

# Top 2 tháng âm nhất theo từng năm, chỉ lấy các tháng thật sự âm
top2_negative_months_each_year = (
    monthly_summary[monthly_summary["Gross_Profit"] < 0]
    .sort_values(["Year", "Gross_Profit"], ascending=[True, True])
    .groupby("Year")
    .head(2)
    .reset_index(drop=True)
)

negative_month_frequency_top2 = (
    top2_negative_months_each_year.groupby(["Month", "Month Name"], as_index=False)
    .agg(
        Frequency_in_Negative_Top2=("Gross_Profit", "size"),
        Total_Gross_Profit=("Gross_Profit", "sum"),
    )
    # Ưu tiên tần suất cao hơn, nếu bằng nhau thì ưu tiên âm hơn
    .sort_values(["Frequency_in_Negative_Top2", "Total_Gross_Profit", "Month"], ascending=[False, True, True])
)

top3_negative_months_most_frequent = negative_month_frequency_top2.head(3)

# =========================
# 3) Thống kê theo quý
# =========================
quarterly_summary = (
    df_daily.groupby(["Year", "Quarter"], as_index=False)
    .agg(
        Revenue=("Revenue", "sum"),
        COGS=("COGS", "sum"),
        Gross_Profit=("Gross Profit", "sum"),
    )
)

# Quý có lợi nhuận gộp cao nhất theo từng năm
top_quarter_each_year = (
    quarterly_summary.sort_values(["Year", "Gross_Profit"], ascending=[True, False])
    .groupby("Year")
    .head(1)
    .reset_index(drop=True)
)

# Top 4 quý có lợi nhuận gộp cao nhất theo từng năm
top4_quarters_each_year = (
    quarterly_summary.sort_values(["Year", "Gross_Profit"], ascending=[True, False])
    .groupby("Year")
    .head(4)
    .reset_index(drop=True)
)

quarter_frequency_top4 = (
    top4_quarters_each_year.groupby(["Quarter"], as_index=False)
    .agg(
        Frequency_in_Top4=("Gross_Profit", "size"),
        Total_Gross_Profit=("Gross_Profit", "sum"),
    )
    .sort_values(["Frequency_in_Top4", "Total_Gross_Profit", "Quarter"], ascending=[False, False, True])
)

top3_quarters_most_frequent = quarter_frequency_top4.head(3)

# Các quý có lợi nhuận gộp âm mạnh nhất
bottom_quarters_gross_profit = (
    quarterly_summary.sort_values("Gross_Profit", ascending=True)
    .head(10)
    .reset_index(drop=True)
)

# =========================
# 4) Thống kê doanh thu thấp
# =========================
bottom_months_revenue = (
    monthly_summary.sort_values("Revenue", ascending=True)
    .head(10)
    .reset_index(drop=True)
)

bottom_quarters_revenue = (
    quarterly_summary.sort_values("Revenue", ascending=True)
    .head(10)
    .reset_index(drop=True)
)

# =========================
# 5) In kết quả
# =========================
pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 20)

print("===== TOP 5 THÁNG CÓ LỢI NHUẬN GỘP CAO NHẤT THEO TỪNG NĂM =====")
display(top5_months_each_year[["Year", "Month Name", "Revenue", "COGS", "Gross_Profit"]])

print("===== TẦN SUẤT XUẤT HIỆN CỦA CÁC THÁNG TRONG TOP 5 =====")
display(month_frequency_top5.rename(columns={
    "Frequency_in_Top5": "Frequency in Top 5",
    "Total_Gross_Profit": "Total Gross Profit"
}))

print("===== TOP 3 THÁNG XUẤT HIỆN NHIỀU NHẤT TRONG TOP 5 =====")
display(top3_months_most_frequent.rename(columns={
    "Frequency_in_Top5": "Frequency in Top 5",
    "Total_Gross_Profit": "Total Gross Profit"
}))

print("===== TOP 3 THÁNG XUẤT HIỆN NHIỀU NHẤT TRONG TOP 2 THÁNG LỢI NHUẬN GỘP ÂM =====")
display(top3_negative_months_most_frequent.rename(columns={
    "Frequency_in_Negative_Top2": "Frequency in Negative Top 2",
    "Total_Gross_Profit": "Total Gross Profit"
}))

print("===== CÁC THÁNG CÓ LỢI NHUẬN GỘP ÂM MẠNH NHẤT =====")
display(bottom_months_gross_profit[["Year", "Month", "Gross_Profit", "Revenue", "COGS"]].assign(
    **{"Month Name": bottom_months_gross_profit["Month"].map(month_map)}
)[["Year", "Month", "Month Name", "Revenue", "COGS", "Gross_Profit"]])

print("===== QUÝ CÓ LỢI NHUẬN GỘP CAO NHẤT THEO TỪNG NĂM =====")
display(top_quarter_each_year[["Year", "Quarter", "Revenue", "COGS", "Gross_Profit"]])

print("===== TẦN SUẤT XUẤT HIỆN CỦA QUÝ TRONG TOP 4 =====")
display(quarter_frequency_top4.rename(columns={
    "Frequency_in_Top4": "Frequency in Top 4",
    "Total_Gross_Profit": "Total Gross Profit"
}))

print("===== TOP 3 QUÝ XUẤT HIỆN NHIỀU NHẤT TRONG TOP 4 =====")
display(top3_quarters_most_frequent.rename(columns={
    "Frequency_in_Top4": "Frequency in Top 4",
    "Total_Gross_Profit": "Total Gross Profit"
}))

print("===== CÁC QUÝ CÓ LỢI NHUẬN GỘP ÂM MẠNH NHẤT =====")
display(bottom_quarters_gross_profit[["Year", "Quarter", "Revenue", "COGS", "Gross_Profit"]])

print("===== CÁC THÁNG CÓ DOANH THU THẤP NHẤT =====")
display(bottom_months_revenue.assign(
    **{"Month Name": bottom_months_revenue["Month"].map(month_map)}
)[["Year", "Month", "Month Name", "Revenue", "COGS", "Gross_Profit"]])

print("===== CÁC QUÝ CÓ DOANH THU THẤP NHẤT =====")
display(bottom_quarters_revenue[["Year", "Quarter", "Revenue", "COGS", "Gross_Profit"]])
```

## 📦 PHẦN 3: PHÂN TÍCH CƠ CẤU DOANH THU & LỢI NHUẬN THEO DANH MỤC

### 🎯 1. Mục tiêu của phần này

- Phát hiện các **Margin Killers** - những nhóm hàng có doanh thu lớn nhưng lợi nhuận thực tế thấp, thậm chí âm, do biên lợi nhuận kém, chi phí đầu vào cao hoặc tỷ lệ trả hàng lớn.
- Tách bạch giữa **quy mô bán hàng** và **hiệu quả sinh lời**, tránh nhầm lẫn giữa “bán được nhiều” với “kinh doanh tốt”.

Nói cách khác, phần này giúp ban điều hành nhìn rõ đâu là danh mục nên ưu tiên mở rộng, và đâu là danh mục cần xem xét lại chiến lược giá, sourcing hoặc vận hành.

---

### 🌳 2. Tại sao chọn biểu đồ Treemap?
Treemap là lựa chọn phù hợp nhất cho bài toán này vì nó cho phép phân tích đồng thời **cấu trúc phân cấp** và **hiệu quả kinh doanh** trên cùng một khung nhìn.

#### 🔍 Phân tích đa tầng Hierarchy
Treemap thể hiện rõ mối quan hệ từ:

- **Danh mục lớn Category**
- xuống **Phân khúc khách hàng Segment**
- và có thể mở rộng sâu hơn nếu cần theo dòng sản phẩm hoặc nhãn hàng

Điều này rất quan trọng với dashboard cấp điều hành, vì nhà quản trị không chỉ muốn biết “cái gì đang bán chạy”, mà còn muốn biết “bán cho ai” và “nhóm nào đang tạo lợi nhuận thật”.

#### 📐 Hai chiều thông tin trong một biểu đồ
Treemap mang lại hai lớp thông tin cùng lúc:

- **Diện tích ô** đại diện cho **Doanh thu** hoặc quy mô đóng góp.
- **Màu sắc ô** đại diện cho **Biên lợi nhuận gộp** hoặc hiệu quả sinh lời.

Nhờ vậy, người xem có thể phát hiện ngay:

- Danh mục vừa lớn vừa hiệu quả
- Danh mục lớn nhưng lợi nhuận yếu
- Danh mục nhỏ nhưng biên lợi nhuận rất tốt
- Danh mục tạo doanh thu cao nhưng đang “ăn mòn” lợi nhuận

Đây là một cách trực quan hóa rất mạnh để tìm ra các điểm nóng trong cơ cấu danh mục.

---

### 🎨 3. Chiến lược màu sắc Blue-to-Red
Hệ màu được thiết kế theo nguyên tắc **Blue-to-Red** để phản ánh trực tiếp chất lượng lợi nhuận.

- **Xanh dương đậm:** Những nhóm hàng có biên lợi nhuận gộp cao, thường trên ngưỡng 30%, thể hiện đây là các danh mục khỏe, có khả năng tạo lợi nhuận tốt và là động lực tăng trưởng chính.
- **Đỏ đậm:** Những nhóm hàng có biên lợi nhuận thấp hoặc âm, cho thấy cần xem lại giá nhập, chi phí vận hành, khuyến mãi hoặc mức độ trả hàng.

Chiến lược màu này giúp dashboard không chỉ đẹp về mặt thị giác mà còn mang tính cảnh báo rõ ràng, hỗ trợ ra quyết định nhanh ngay ở tầng quản trị.

### 🧠 4. Câu hỏi chẩn đoán Diagnostic Questions
Phần này không dừng ở việc mô tả cơ cấu danh mục, mà hướng tới chẩn đoán nguyên nhân đằng sau hiệu quả kinh doanh.

- Liệu phân khúc **Premium** có thực sự mang lại lợi nhuận cao hơn **Standard** như kỳ vọng, hay chỉ tạo ra doanh thu lớn nhưng bị bào mòn bởi chi phí?
- Nhóm hàng nào đang chịu ảnh hưởng lớn nhất từ việc **trả hàng**, khiến doanh thu danh nghĩa cao nhưng lợi nhuận thực tế suy giảm?
- Danh mục nào đang là **Hero Categories** thực sự, và danh mục nào đang âm thầm trở thành **Margin Killers** trong toàn bộ cơ cấu kinh doanh?
- Có tồn tại trường hợp một phân khúc nhìn rất mạnh về quy mô nhưng lại có biên lợi nhuận thấp hơn mức chấp nhận được hay không?

### ✅ Kết luận ngắn
Treemap là công cụ phù hợp nhất để nhìn ra sự thật phía sau doanh thu: **ai đang tạo ra tiền, ai đang phá lợi nhuận, và lợi nhuận đang đến từ đâu**.  
Phần này sẽ giúp doanh nghiệp chuyển từ tư duy “bán nhiều là tốt” sang tư duy “bán đúng danh mục mới là tăng trưởng bền vững”.

```python
# Treemap: Cơ cấu doanh thu & hiệu quả lợi nhuận theo danh mục

import numpy as np
import pandas as pd
import plotly.express as px

# =========================
# 1) Chuẩn bị dữ liệu
# =========================

# Chuẩn hóa tên cột nếu cần
df_products = dfs["products"]
df_oi = df_order_items.copy()
df_pr = df_products.copy()

df_oi.columns = df_oi.columns.str.strip()
df_pr.columns = df_pr.columns.str.strip()

# Gom các tên cột phổ biến về dạng có thể dùng được
rename_map_products = {}
if "Category" in df_pr.columns and "category" not in df_pr.columns:
    rename_map_products["Category"] = "category"
if "Segment" in df_pr.columns and "segment" not in df_pr.columns:
    rename_map_products["Segment"] = "segment"
if "Price" in df_pr.columns and "price" not in df_pr.columns:
    rename_map_products["Price"] = "price"
if "COGS" in df_pr.columns and "cogs" not in df_pr.columns:
    rename_map_products["COGS"] = "cogs"

df_pr = df_pr.rename(columns=rename_map_products)

rename_map_oi = {}
if "Quantity" in df_oi.columns and "quantity" not in df_oi.columns:
    rename_map_oi["Quantity"] = "quantity"
if "Unit_Price" in df_oi.columns and "unit_price" not in df_oi.columns:
    rename_map_oi["Unit_Price"] = "unit_price"
if "Product_ID" in df_oi.columns and "product_id" not in df_oi.columns:
    rename_map_oi["Product_ID"] = "product_id"
if "Order_ID" in df_oi.columns and "order_id" not in df_oi.columns:
    rename_map_oi["Order_ID"] = "order_id"

df_oi = df_oi.rename(columns=rename_map_oi)

# Merge deep giữa order_items và products
df_treemap = df_oi.merge(
    df_pr[["product_id", "category", "segment", "price", "cogs"]],
    on="product_id",
    how="left"
)

# Nếu order_items có unit_price thì ưu tiên dùng unit_price,
# còn không thì dùng price từ products
if "unit_price" not in df_treemap.columns:
    df_treemap["unit_price"] = df_treemap["price"]
else:
    df_treemap["unit_price"] = df_treemap["unit_price"].fillna(df_treemap["price"])

# Ép kiểu số an toàn
for col in ["quantity", "unit_price", "cogs"]:
    if col in df_treemap.columns:
        df_treemap[col] = pd.to_numeric(df_treemap[col], errors="coerce")

# Tính toán theo từng dòng
df_treemap["Doanh thu"] = df_treemap["quantity"] * df_treemap["unit_price"]
df_treemap["Lợi nhuận gộp"] = df_treemap["Doanh thu"] - (df_treemap["quantity"] * df_treemap["cogs"])

# Làm sạch dữ liệu nhóm
df_treemap["category"] = df_treemap["category"].fillna("Không xác định")
df_treemap["segment"] = df_treemap["segment"].fillna("Không xác định")

# Groupby theo category và segment
df_treemap_grouped = (
    df_treemap
    .groupby(["category", "segment"], as_index=False)
    .agg(
        Doanh_thu=("Doanh thu", "sum"),
        Loi_nhuan_gop=("Lợi nhuận gộp", "sum")
    )
)

# Tính biên lợi nhuận gộp
df_treemap_grouped["Bien_loi_nhuan_gop"] = np.where(
    df_treemap_grouped["Doanh_thu"] != 0,
    df_treemap_grouped["Loi_nhuan_gop"] / df_treemap_grouped["Doanh_thu"],
    np.nan
)

# =========================
# 2) Vẽ Treemap
# =========================

fig = px.treemap(
    df_treemap_grouped,
    path=["category", "segment"],
    values="Doanh_thu",
    color="Bien_loi_nhuan_gop",
    color_continuous_scale=px.colors.diverging.RdBu,  # đảo ngược để xanh = cao, đỏ = thấp
    color_continuous_midpoint=0.138, # baseline la 13.8%
    custom_data=["Doanh_thu", "Loi_nhuan_gop", "Bien_loi_nhuan_gop"]
)

fig.update_traces(
    textinfo="label+value",
    texttemplate="<b>%{label}</b><br>%{value:,.0f} VNĐ",
    hovertemplate=(
        "<b>%{label}</b><br>"
        "Doanh thu: %{customdata[0]:,.0f} VNĐ<br>"
        "Lợi nhuận gộp: %{customdata[1]:,.0f} VNĐ<br>"
        "Biên lợi nhuận gộp: %{customdata[2]:.1%}<extra></extra>"
    ),
    marker=dict(line=dict(color="white", width=2))
)

fig.update_layout(
    title={
        "text": "CƠ CẤU DOANH THU & HIỆU QUẢ LỢI NHUẬN THEO DANH MỤC",
        "x": 0.5,
        "xanchor": "center"
    },
    template="plotly_white",
    height=600,
    margin=dict(t=80, l=20, r=20, b=20)
)

fig.show()
```

## 🌐 PHẦN 4: PHÂN TÍCH HIỆU QUẢ CHUYỂN ĐỔI (LƯU LƯỢNG WEB VS. ĐƠN HÀNG)

### 🎯 1. Mục tiêu của phần này
Phần này được thiết kế để **kết nối dữ liệu Vận hành (Web Traffic)** với dữ liệu **Giao dịch (Orders)** nhằm đánh giá hiệu quả của toàn bộ phễu bán hàng từ truy cập đến chốt đơn.

Mục tiêu trọng tâm là trả lời câu hỏi:

- **Lưu lượng truy cập tăng có thực sự giúp chốt đơn tốt hơn không?**
- Hay website đang có nhiều traffic nhưng chất lượng traffic thấp, khiến tỷ lệ chuyển đổi không cải thiện tương xứng?

Nói cách khác, phần này giúp ban điều hành nhìn rõ:
- Traffic đang đến từ đâu,
- Traffic đó có đủ chất lượng hay không,
- Và doanh nghiệp đang chuyển đổi khách truy cập thành đơn hàng hiệu quả đến mức nào.

---

### 📊 2. Các chỉ số đo lường cốt lõi
#### 🔹 Lượt truy cập Visits
Tổng số lượt tương tác trên website trong kỳ phân tích. Đây là chỉ số phản ánh quy mô tiếp cận và mức độ quan tâm của người dùng.

#### 🔹 Số đơn hàng Orders
Tổng số đơn hàng thành công phát sinh trong cùng kỳ. Đây là chỉ số phản ánh kết quả thực tế của hoạt động bán hàng.

#### 🔹 Tỷ lệ Chuyển đổi Conversion Rate - CR%
Chỉ số thể hiện mức độ hiệu quả của phễu bán hàng, được tính theo công thức:

- **CR% = (Số đơn hàng / Visits) × 100**

CR% càng cao, phễu chuyển đổi càng hiệu quả. Ngược lại, nếu Visits tăng nhưng CR% giảm, đó là dấu hiệu cho thấy chất lượng traffic có thể đang suy yếu hoặc trải nghiệm mua hàng chưa đủ tốt.

---

### 🛠️ 3. Kỹ thuật phân tích & trực quan
#### 📈 Biểu đồ kết hợp Dual-Axis Chart
Phần này nên sử dụng **biểu đồ hai trục** để hiển thị đồng thời:

- **Trục chính:** Quy mô hoạt động, gồm **Visits** và **Orders**
- **Trục phụ:** Hiệu suất chuyển đổi, gồm **CR%**

Cách thể hiện này giúp người xem vừa nhìn được quy mô traffic, vừa đánh giá được chất lượng chuyển đổi trong cùng một khung thời gian.

#### 📉 Làm mượt bằng MA7
Dữ liệu web traffic và đơn hàng theo ngày thường có dao động mạnh do:
- Khác biệt giữa ngày thường và cuối tuần,
- Hiệu ứng chiến dịch quảng cáo ngắn hạn,
- Biến động mùa vụ hoặc sự kiện đặc biệt.

Vì vậy, áp dụng **MA7 (Moving Average 7 ngày)** là cần thiết để:
- Loại bỏ nhiễu ngắn hạn,
- Làm rõ xu hướng thực,
- Nhìn được mối tương quan thật giữa traffic và đơn hàng.

---

### 🎨 4. Logic diễn giải trực quan
#### 🔵 Visits và Orders
Dùng màu xanh dương để biểu thị:
- Dòng traffic đổ vào website,
- Quy mô hoạt động,
- Động lực tăng trưởng của kênh online.

#### 🟢 CR%
Dùng màu xanh lá hoặc xanh ngọc để biểu thị:
- Hiệu quả chuyển đổi,
- Chất lượng phễu bán hàng,
- Mức độ “ra đơn” từ traffic đầu vào.

#### ⚠️ Tín hiệu cảnh báo
Nếu:
- **Visits tăng mạnh nhưng Orders không tăng tương ứng**, hoặc
- **CR% giảm sâu trong giai đoạn traffic cao**,

thì đây là dấu hiệu cho thấy:
- Chất lượng traffic có thể không phù hợp,
- Quảng cáo đang kéo sai đối tượng,
- Hoặc website có vấn đề về trải nghiệm, tốc độ, nội dung sản phẩm, giá bán hay thanh toán.

### 🧠 5. Câu hỏi chẩn đoán Diagnostic Questions
Phần này không chỉ mô tả dữ liệu, mà còn hướng tới chẩn đoán nguyên nhân phía sau hiệu suất chuyển đổi.

- Có giai đoạn nào **Visits tăng vọt nhưng CR% lại giảm sâu** không?
  - Nếu có, đây có thể là dấu hiệu của lỗi kỹ thuật trên web, tracking sai, hoặc chiến dịch quảng cáo đang kéo sai tệp khách hàng.

- **CR% có ổn định qua các mùa cao điểm như Tết hay không?**
  - Nếu CR% giữ được nhịp ổn định hoặc cải thiện trong giai đoạn traffic tăng mạnh, đó là dấu hiệu của một phễu bán hàng khỏe.
  - Ngược lại, nếu traffic tăng nhưng CR% giảm, doanh nghiệp cần rà soát lại chất lượng nguồn vào và hành trình mua hàng.

- Traffic tăng đến từ **tăng nhận biết** hay chỉ là **tăng truy cập không chất lượng**?
- Số đơn hàng tăng có đi cùng với việc cải thiện hiệu suất chuyển đổi hay không?

### ✅ 6. Kết luận ngắn
Phần này giúp doanh nghiệp chuyển từ câu hỏi “website có nhiều người vào không?” sang câu hỏi quan trọng hơn: **“traffic đó có tạo ra doanh thu thật không?”**

Đây là lớp phân tích then chốt để đánh giá chất lượng tăng trưởng kênh online, từ đó tối ưu:
- Nội dung quảng cáo,
- Chất lượng landing page,
- Trải nghiệm mua hàng,
- Và hiệu quả chuyển đổi toàn phễu.

```python
# =========================
# 1) Tiền xử lý dữ liệu
# =========================

# Chuẩn hóa tên cột nếu cần
df_orders = dfs["orders"].copy()
df_web_traffic = dfs["web_traffic"].copy()

df_orders.columns = df_orders.columns.str.strip()
df_web_traffic.columns = df_web_traffic.columns.str.strip()

# Nếu cột ngày đang có tên khác, tự động map về "date"
if "date" not in df_orders.columns:
    if "order_date" in df_orders.columns:
        df_orders = df_orders.rename(columns={"order_date": "date"})
    elif "Date" in df_orders.columns:
        df_orders = df_orders.rename(columns={"Date": "date"})

if "date" not in df_web_traffic.columns:
    if "Date" in df_web_traffic.columns:
        df_web_traffic = df_web_traffic.rename(columns={"Date": "date"})

# Đảm bảo kiểu dữ liệu ngày
df_orders["date"] = pd.to_datetime(df_orders["date"], errors="coerce")
df_web_traffic["date"] = pd.to_datetime(df_web_traffic["date"], errors="coerce")

# Đảm bảo cột unique_visitors là số
if "unique_visitors" not in df_web_traffic.columns:
    raise ValueError("df_web_traffic cần có cột 'unique_visitors'.")

df_web_traffic["unique_visitors"] = pd.to_numeric(df_web_traffic["unique_visitors"], errors="coerce")
df_orders["order_id"] = pd.to_numeric(df_orders["order_id"], errors="coerce")

# Groupby theo ngày
df_orders_daily = (
    df_orders.dropna(subset=["date", "order_id"])
    .groupby("date", as_index=False)
    .agg(so_don_hang=("order_id", "count"))
)

df_traffic_daily = (
    df_web_traffic.dropna(subset=["date", "unique_visitors"])
    .groupby("date", as_index=False)
    .agg(luot_truy_cap=("unique_visitors", "sum"))
)

# Merge hai bảng
df_conv = pd.merge(df_traffic_daily, df_orders_daily, on="date", how="inner")

# Tính tỷ lệ chuyển đổi
df_conv["ty_le_chuyen_doi"] = np.where(
    df_conv["luot_truy_cap"] != 0,
    df_conv["so_don_hang"] / df_conv["luot_truy_cap"] * 100,
    np.nan
)

# Sắp xếp theo thời gian trước khi rolling
df_conv = df_conv.sort_values("date").reset_index(drop=True)

# MA7
df_conv["so_don_hang_ma7"] = df_conv["so_don_hang"].rolling(window=7, min_periods=1).mean()
df_conv["luot_truy_cap_ma7"] = df_conv["luot_truy_cap"].rolling(window=7, min_periods=1).mean()
df_conv["cr_ma7"] = df_conv["ty_le_chuyen_doi"].rolling(window=7, min_periods=1).mean()

# =========================
# 2) Biểu đồ Dual Axis
# =========================

fig = go.Figure()

# Trace 1: unique_Visitors nền
fig.add_trace(
    go.Bar(
        x=df_conv["date"],
        y=df_conv["luot_truy_cap"],
        name="Lượt truy cập",
        marker=dict(color="rgba(255, 175, 175, 0.8)"),
        hovertemplate="%{x|%d/%m/%Y}<br>Lượt truy cập: %{y:,.0f}<extra></extra>",
    )
)

# Trace 2: Orders MA7 - trục trái
fig.add_trace(
    go.Scatter(
        x=df_conv["date"],
        y=df_conv["so_don_hang_ma7"],
        mode="lines",
        name="Số đơn hàng (MA7)",
        line=dict(color="rgba(30, 90, 180, 1)", width=3),
        hovertemplate="%{x|%d/%m/%Y}<br>Số đơn hàng (MA7): %{y:,.2f}<extra></extra>",
        yaxis="y",
    )
)

# Trace 3: CR MA7 - trục phải
fig.add_trace(
    go.Scatter(
        x=df_conv["date"],
        y=df_conv["cr_ma7"],
        mode="lines",
        name="Tỷ lệ chuyển đổi (MA7)",
        line=dict(color="rgba(255, 140, 0, 1)", width=3),
        hovertemplate="%{x|%d/%m/%Y}<br>CR (MA7): %{y:,.2f}%<extra></extra>",
        yaxis="y2",
    )
)

# Cấu hình trục
fig.update_layout(
    title={
        "text": "TƯƠNG QUAN LƯU LƯỢNG TRUY CẬP VÀ HIỆU QUẢ CHUYỂN ĐỔI (MA7)",
        "x": 0.5,
        "xanchor": "center",
    },
    template="plotly_white",
    height=650,
    margin=dict(l=70, r=70, t=90, b=70),
    legend=dict(
        title="Chú thích",
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
    ),
    barmode="overlay",
)

fig.update_xaxes(
    title_text="Thời gian",
    rangeslider=dict(visible=True),
    showgrid=False,
)

fig.update_yaxes(
    title_text="Lượt truy cập / Số đơn hàng",
    showgrid=True,
    gridcolor="rgba(0,0,0,0.08)",
)

fig.update_layout(
    yaxis2=dict(
        title="Tỷ lệ chuyển đổi (%)",
        overlaying="y",
        side="right",
        tickformat=".2f",
        ticksuffix="%",
        showgrid=False,
    )
)

fig.show()
```
